```python
import cryptocompare
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from datetime import datetime

%matplotlib notebook

plt.style.use('seaborn')
```

```python
#get price a cryptocurrency in a fiat money
def crypto_price(crypto,fiat):
    return cryptocompare.get_price(crypto,fiat)[crypto][fiat]

print(crypto_price("BTC","EUR"))
```

```python
#get name a cryptocurrency
def crypto_name(crypto):
  return cryptocompare.get_coin_list()[crypto]["FullName"]

print(crypto_name("BTC"))
```

```python
# Animation

x = [] # date
y = [] # Price
crypto="BTC" # ETH, BNB ...
fiat = "EUR" # USD

def animation(i):
  x.append(datetime.now())
  y.append(crypto_price(crypto,fiat))

  plt.title(crypto_name(crypto) + " Price Animation")
  plt.xlabel("Time")
  plt.ylabel("Price")
  #plt.ylabel("Price", rotation=0 )

  # plotting according the date
  plt.plot_date(x,y, linestyle="-", markersize=0.2)
   
  # (optional) use for adjusting graphic 
  plt.tight_layout()


# call FuncAnimation function which calls "animation" every 1000 millisec
ani = FuncAnimation(plt.gcf(), animation, interval=1000)

plt.show()
```
